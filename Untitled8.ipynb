```python
import pandas as pd
import numpy as np

# If already uploaded manually
df = pd.read_csv("data.csv", encoding="latin1")

# Clean column names
df.columns = df.columns.str.strip()

print(df.columns)
```

```python
# Drop missing values
df = df.dropna(subset=["no2"])

x = df["no2"].values

# Remove extreme outliers
x = x[x < np.percentile(x, 99)]

# Your roll number parameters
a_r = 3.0
b_r = 0.9

# Transformation
z = x + a_r * np.sin(b_r * x)
z = z.reshape(-1,1)

print("Min z:", z.min())
print("Max z:", z.max())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
z_scaled = scaler.fit_transform(z)

print("Min scaled:", z_scaled.min())
print("Max scaled:", z_scaled.max())
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

z_tensor = torch.tensor(z_scaled, dtype=torch.float32).to(device)
```

```python
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(1, 32),
            nn.ReLU(),
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.Linear(64, 1),
            nn.Sigmoid()   # VERY IMPORTANT
        )

    def forward(self, x):
        return self.model(x)


class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(1, 64),
            nn.LeakyReLU(0.2),
            nn.Linear(64, 32),
            nn.LeakyReLU(0.2),
            nn.Linear(32, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.model(x)
```

```python
G = Generator().to(device)
D = Discriminator().to(device)

criterion = nn.BCELoss()

optimizer_G = optim.Adam(G.parameters(), lr=0.0001)
optimizer_D = optim.Adam(D.parameters(), lr=0.0001)
```

```python
epochs = 5000
batch_size = 128

for epoch in range(epochs):

    idx = np.random.randint(0, z_scaled.shape[0], batch_size)
    real_samples = z_tensor[idx]

    noise = torch.randn(batch_size, 1).to(device)
    fake_samples = G(noise)

    real_labels = torch.ones(batch_size, 1).to(device)
    fake_labels = torch.zeros(batch_size, 1).to(device)

    # Train D
    D_loss_real = criterion(D(real_samples), real_labels)
    D_loss_fake = criterion(D(fake_samples.detach()), fake_labels)
    D_loss = D_loss_real + D_loss_fake

    optimizer_D.zero_grad()
    D_loss.backward()
    optimizer_D.step()

    # Train G
    noise = torch.randn(batch_size, 1).to(device)
    fake_samples = G(noise)

    G_loss = criterion(D(fake_samples), real_labels)

    optimizer_G.zero_grad()
    G_loss.backward()
    optimizer_G.step()

    if epoch % 1000 == 0:
        print(f"Epoch {epoch} | D Loss: {D_loss.item():.4f} | G Loss: {G_loss.item():.4f}")
```

```python
G.eval()

noise = torch.randn(10000, 1).to(device)
generated_scaled = G(noise).detach().cpu().numpy()

# IMPORTANT
generated = scaler.inverse_transform(generated_scaled)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

real_z = z.flatten()

plt.figure(figsize=(8,5))

sns.kdeplot(real_z, label="True PDF (Data)")
sns.kdeplot(generated.flatten(), label="Learned PDF (GAN)", linestyle="--")

plt.title("PDF Learning using GAN")
plt.xlabel("z")
plt.ylabel("Density")
plt.legend()
plt.show()
```
